```python
import numpy as np
import pandas as pd

columns = ['user_id', 'item_id', 'rating', 'timestamp', 'title', 'people', 'country', 'genre']
dataset = pd.read_csv("naver_movie_dataset.csv", names=columns)
```

```python
dataset1 = dataset[['item_id','title']]
dataset1.drop_duplicates(inplace=True, subset='item_id', keep="last")
products_dict = dataset1.groupby('item_id')['title'].apply(list).to_dict()
products_dict
```

```python
num_user = dataset['user_id'].value_counts()
# df 형식으로 변환
num_user = pd.DataFrame({'user_id':num_user.index, 'counts':num_user.values})
num_user
```

```python
# 유일한 VOD ID 값
num_vod = dataset['item_id'].value_counts()
# df 형식으로 변환
num_vod = pd.DataFrame({'item_id':num_vod.index, 'counts':num_vod.values})
num_vod
# 491539 중 시청한 VOD 수 : 1034
```

```python
# 시청 횟수가 5회 이상인 유저 수
num_user_5 = num_user[num_user['counts']>4]
num_user_5
# 약 5만명의 user 중 시청 횟수 5이상은 2만명
```

```python
# 시청 횟수가 5회 이상인 VOD 
num_vod_5 = num_vod[num_vod['counts']>4]
num_vod_5
# 1034개 VOD 중 시청 횟수 5이상은 994개
```

```python
# 5회 이상 시청한 이력만 추출하기
check = dataset['user_id'].isin(num_user_5['user_id'])
df = dataset[dataset['user_id'].isin(num_user_5['user_id'])]
df
```

```python
df = df.sort_values(by='user_id')
#df.to_csv("data_user.csv")
df
```

```python
df_list = df.groupby('user_id')['item_id'].apply(list)
# df_list.to_csv("user_vod.csv")
df_list
```

```python
vec = df_list.values
```

```python
vec
```

```python
from gensim.models.word2vec import Word2Vec
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
```

```python
def plot_2d_graph(vocabs, xs, ys):
    plt.figure(figsize=(32, 24))
    plt.scatter(xs, ys, marker = 'o')
    for i, v in enumerate(vocabs):
        plt.annotate(v, xy=(xs[i], ys[i]))
    # plt 저장
    plt.savefig('w2v_result.png', dpi=100)
```

```python
model = Word2Vec(vec,         # 리스트 형태의 데이터
                 sg=1,         # 0: CBOW, 1: Skip-gram
                 vector_size=100,     # 벡터 크기
                 window=2,     # 고려할 앞뒤 폭(앞뒤 3단어)
                 min_count=3,  # 사용할 단어의 최소 빈도(3회 이하 단어 무시)
                )    # 동시에 처리할 작업 수(코어 수와 비슷하게 설정)
```

```python
model = Word2Vec(window = 10, sg = 1, hs = 0,
                 negative = 10, # for negative sampling
                 alpha=0.03, min_alpha=0.0007,
                 seed = 14)

model.build_vocab(vec, progress_per=200)

model.train(vec, total_examples = model.corpus_count, 
            epochs=10, report_delay=1)
```

```python
def similar_products(v, n = 6):
    
    # extract most similar products for the input vector
    ms = model.similar_by_vector(v, topn= n+1)[1:]
    
    # extract name and similarity score of the similar products
    new_ms = []
    for j in ms:
        pair = (products_dict[j[0]][0], j[1])
        new_ms.append(pair)
        
    return new_ms        
```

```python
for index, word in enumerate(model.wv.index_to_key):
    if index == 10:
        break
    print(f"word #{index}/{len(model.wv.index_to_key)} is {word}")
```

```python
def recommendation(num):
    rec = model.wv.most_similar(num)
    print("영화 추천 리스트\n")
    print("선택한 영화 : ", products_dict[num])
    for i in rec:
        print(products_dict[i[0]])
```

```python
recommendation(5)
```
